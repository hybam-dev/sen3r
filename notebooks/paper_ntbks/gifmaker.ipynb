```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
efrl1 = mpimg.imread('C:\\Users\\david\\Pictures\\beamer\\S3A_OL_1_EFR____20191104T135002.jpg')
wfrl2 = mpimg.imread('C:\\Users\\david\\Pictures\\beamer\\S3A_OL_2_WFR____20191104T135002.jpg')

imglist = [efrl1, wfrl2]
```

```python
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
```

```python
class LoopingPillowWriter(PillowWriter):
    def finish(self):
        self._frames[0].save(
            self._outfile, save_all=True, append_images=self._frames[1:], bbox_inches='tight',
            duration=int(1000 / self.fps), loop=0)
```

```python
plt.rcParams['figure.figsize'] = [12, 8]
fig = plt.figure()
ims = []

for f in imglist:
    frame = plt.imshow(f)
    ims.append([frame])
```

```python
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat_delay=500)

writer = PillowWriter(fps=1)
# ani.save("demo2.gif", writer=writer)
ani.save('efr_wfr.gif', writer=LoopingPillowWriter(fps=1)) 

plt.show()
```
